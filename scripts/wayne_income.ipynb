## Relationship between Income and Condition of the Road in Wayne, MI


```python
import pandas as pd
```

```python
# Read in the files from geo sampling that are fed to Mturk
wayne   = pd.read_csv("../data/google_street_view_metadata/wayne2-askgeo.csv")

# Read in the Mturk data
wayne_mturk   = pd.read_csv("../data/mturk/wayne_mturk_2018_06_11_18.csv")
```

```python
wayne_mturk.head()
```

```python
# Left join original data with Mturk
wayne_all   = wayne.merge(wayne_mturk, how = 'left', left_on = 'url_img0', right_on = 'Input.url_img0')
```

```python
# Number of rows in the final data (confirming)
print(wayne_all.shape[0])
```

```python
# Let's subset the data to relevant cols
small_subset = ('osm_type',
                'date',
                'UsTract2010_AcsIncomePerCapita',
                'Answer.Do you see any of the following?',
                'Answer.Does the road have any...?',
                'Answer.If there are any road markings are they clear?')
wayne_small   = wayne_all.loc[:, small_subset]
```

```python
# Recoding for convenience and clarity

# Let's tally proportion of roads with potholes and cracks
wayne_small['cracks']   = wayne_small['Answer.Does the road have any...?'].str.contains('Cracks', na = False)
wayne_small['potholes']   = wayne_small['Answer.Does the road have any...?'].str.contains('Potholes', na = False)
wayne_small['clear_road_markings']   = wayne_small['Answer.If there are any road markings are they clear?'] == 'Yes'
wayne_small['roads_w_markings']      = wayne_small['Answer.If there are any road markings are they clear?'].str.contains('Yes|No', na = False)
wayne_small['litter']   = wayne_small['Answer.Do you see any of the following?'].str.contains('Litter', na = False)
wayne_small['paved_sidewalk']   = wayne_small['Answer.Do you see any of the following?'].str.contains('sidewalk', na = False)

# Let's subset on relevant cols.
fin_cols = ('osm_type', 'date', 'UsTract2010_AcsIncomePerCapita', 'potholes', 'cracks', 'clear_road_markings', 'roads_w_markings', 'litter', 'paved_sidewalk')
wayne_final = wayne_small.loc[:, fin_cols]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Drop NAs
wayne_finals = wayne_final.loc[:, ('UsTract2010_AcsIncomePerCapita', 'potholes', 'cracks')]
wayne_finals.dropna(inplace=True)
print(wayne_finals.describe())

X_train    = pd.qcut(wayne_finals['UsTract2010_AcsIncomePerCapita'], 5, labels=["5k_12k", "12k_17k", "17_23k", "23k_29k", "29k_83k"])
X_train_fin = pd.get_dummies(X_train)
X_train_finn = X_train_fin.loc[:, ("12k_17k", "17_23k", "23k_29k", "29k_83k")]

y_potholes = wayne_finals[['potholes']]
y_cracks = wayne_finals[['cracks']]

print("\nPotholes:\n")
lm = LinearRegression()
lm.fit(X_train_finn, y_potholes)

for idx, col_name in enumerate(X_train_finn.columns):
    print("The coefficient for {} is {}".format(col_name, lm.coef_[0][idx].round(2)))

print("\nCracks:\n")
lm = LinearRegression()
lm.fit(X_train_finn, y_cracks)

for idx, col_name in enumerate(X_train_finn.columns):
    print("The coefficient for {} is {}".format(col_name, lm.coef_[0][idx].round(2)))
```
